```python
ds = pd.read_csv('C:\\Datasets\\news.csv')
```

```python
ds.head(1)
```

```python
ds=ds.drop(['Unnamed: 0'], axis=1)
```

```python
ds['label'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
ds['label']=LE.fit_transform(ds['label'])
```

```python
ds = ds[0:1000]
```

```python
ds.columns
```

```python
X = ds[['title', 'text']].values
```

```python
X
```

```python
Y = ds[['label']].values
```

```python
Y
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
mat_body = cv.fit_transform(X[:,1]).todense()
```

```python
mat_body
```

```python
cv_head = CountVectorizer(max_features= 5000)
mat_head = cv_head.fit_transform(X[:,1]).todense()
```

```python
mat_head
```

```python
X_mat = np.hstack((mat_head, mat_body))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_mat, Y,test_size = 0.2, random_state = 0 )
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
y_pred = dtc.fit(X_train, Y_train)
```

```python
dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Y_pred = dtc.predict(X_test)
```

```python
accuracy_score(Y_pred, Y_test)
```
